```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
banana_data = pd.read_csv('banana_quality.csv')
```

```python
banana_data.head()
```

```python
banana_data.tail()
```

```python
banana_data.describe()
```

```python
X = banana_data.drop(columns='Quality', axis=1)
Y = banana_data['Quality']
```

```python
X
```

```python
print(X)
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, Y_train.shape)
```

```python
print(Y.shape, Y_train.shape, X_train.shape)
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

```python
X_test_prediction = classifier.predict(X_test)
```

```python
test_acc = accuracy_score(X_test_prediction, Y_test)
```

```python
print(test_acc)
```
